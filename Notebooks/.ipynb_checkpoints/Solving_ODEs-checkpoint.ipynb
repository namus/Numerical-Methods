# Solving ODEs

$dy/dx = y$

IVP: $y(0) = 1$

Analytical solution: 
$y(x) = e^{x}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(func, y0, start=0, stop=1, h=0.1):
    N = int((stop-start)/h)
    x = np.linspace(start,stop,N+1)
    y = np.empty_like(x)
    y[0] = y0
    for i in range(N):
        y[i+1] = y[i] + h*func(x[i],y[i])
    return x, y
```

```python
def func(x,y):
    return y
```

```python
solution = euler(func, 1, 0, 10, 0.1)
```

```python
x = solution[0]
y = solution[1]
analytical = np.exp(x)
```

```python
plt.plot(x, y, 'r-', label='Euler')
plt.plot(x, analytical, 'b--', label='Analytical')
plt.legend()
```

```python
abs(analytical[-1] - y[-1])
```

```python
def RK2(func, y0, start=0, stop=1, h=0.1):
    N = int((stop-start)/h)
    x = np.linspace(start,stop,N+1)
    y = np.empty_like(x)
    y[0] = y0
    for i in range(N):
        #y[i+1] = y[i] + h*0.5*(func(x[i],y[i]) +  func(x[i+1], y[i] + h*func(x[i],y[i])))
        K1 = h*func(x[i],y[i])
        K2 = h*func(x[i] + h, y[i] + K1)
        y[i+1] = y[i] + 0.5*(K1 + K2)
    return x, y
```

```python
def RK4(func, y0, start=0, stop=1, h=0.1):
    N = int((stop-start)/h)
    x = np.linspace(start,stop,N+1)
    y = np.empty_like(x)
    y[0] = y0
    for i in range(N):
        K1 = h*func(x[i],y[i])
        K2 = h*func(x[i] + 0.5*h, y[i] + 0.5*K1)
        K3 = h*func(x[i] + 0.5*h, y[i] + 0.5*K2)
        K4 = h*func(x[i] + h, y[i] + K3)
        y[i+1] = y[i] + (K1 + 2*K2 + 2*K3 + K4)/6
    return x, y
```

```python
x, y = euler(func, 1, 0, 10, 0.1)
plt.plot(x, y, 'r-', label='Euler')
x, y = RK2(func, 1, 0, 10, 0.1)
plt.plot(x, y, 'b-', label='RK2')
x, y = RK4(func, 1, 0, 10, 0.1)
plt.plot(x, y, 'g-', label='RK4')
analytical = np.exp(x)
plt.plot(x, analytical, 'k--', label='Analytical')
plt.legend()
```

```python
x, y = euler(func, 1, 0, 10, 0.1)
analytical = np.exp(x)
abs(analytical[-1] - y[-1])
```

# Molecular Dynamics using Velocity Verlet

$$dx/dt = v$$
$$dv/dt = -(k/m) x = -x$$

$$x(0) = 10$$
$$v(0) = 0$$

```python
STEPS = 100
dt = 0.1
```

```python
start = 0
end = STEPS * dt
t = np.linspace(start, end, STEPS+1)

x = np.zeros(STEPS+1)
v = np.empty_like(x)

pe = np.empty_like(x)
ke = np.empty_like(x)

x[0] = 10
v[0] = 0

pe[0] = 0.5*x[0]**2
ke[0] = 0.5*v[0]**2
```

```python
for step in range(STEPS):
    x[step+1] = x[step] + v[step]*dt - 0.5*x[step]*dt**2
    v[step+1] = v[step] - 0.5*(x[step] + x[step+1])*dt
    pe[step+1] = 0.5*x[step]**2
    ke[step+1] = 0.5*v[step]**2
# Total energy
te = pe + ke
```

```python
plt.plot(t, x, label='x')
plt.plot(t, v, label='v')
plt.legend()
```

```python
plt.plot(t, pe, label='PE')
plt.plot(t, ke, label='KE')
plt.plot(t, te, label='TE')
plt.legend()
```

```python
plt.plot(t, te, label='TE')
```

```python
STEPS = 100
dt = 0.1

start = 0
end = STEPS * dt
t = np.linspace(start, end, STEPS+1)

a = np.zeros(STEPS+1)
b = np.empty_like(x)
c = np.empty_like(x)

a[0] = 1
b[0] = 0
c[0] = 0

k1 = 0.5
k2 = 0.1
```

```python
for step in range(STEPS):
    a[step+1] = a[step] - dt*k1*a[step] 
    b[step+1] = b[step] + dt*(k1*a[step] - k2*b[step]) 
    c[step+1] = c[step] + dt*k2*b[step] 
```

```python
plt.plot(t, a, label='[A]')
plt.plot(t, b, label='[B]')
plt.plot(t, c, label='[C]')
plt.legend()
```
